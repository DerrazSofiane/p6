```python
def read_yelp_dataset(json_file:str):
    """
    Lis un fichier json et retourne les données sous forme
    de dataframe.
    """
    dataset = []
    for line in open(f'data/{json_file}', encoding="utf8"):
        dataset.append(json.loads(line))
        
    return pd.DataFrame.from_records(dataset)
```

```python
def read_big_yelp_dataset(json_file:str, iter_limit:int):
    records = []
    with open(f'data/{json_file}', encoding="utf8") as fl:
        for i, line in enumerate(fl):
            records.append(json.loads(line))
            if i+1 >= iter_limit:
                break

    dataset = pd.DataFrame(records)
    records = [] # Pour optimiser la mémoire
    return dataset
```

#### Filtre sur les mauvais commentaires

Nous prenons les commentaires pour lesquelles les notes sont supérieures à 3.

```python
bad_reviews = data.loc[data['stars'] < 3]
bad_reviews
```

```python
text_sample = bad_reviews.text.iloc[2]
text_sample
```

**Nous pouvons voir que les données ci-dessus contiennent des éléments inutiles tels que la ponctuaction, des caractères spéciaux ou des mots (à, un, le, la...), que nous allons les supprimer.**

#### Suppression de la ponctuation et des caractères spéciaux

Commençons par la supression des caractères spéciaux.

```python
# Suppression de la ponctuation et des espaces
# '\w+' correspond à tout caractère égal à [a-zA-Z0-9_]
clean_text = []

tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(text_sample) # Renvoie une liste pour chaque mot
# On en profite au passage pour filtrer les mots dont la longueur est inférieure à 2
for token in tokens:
    if len(token)>2:
        clean_text.append(token.lower())

clean_text
```

#### Suppression des stop-words

Prochaine étape : supression des stop word (le, la, les...)

```python
english_stopwords = stopwords.words('english')
filtre_stop = lambda text: [word for word in clean_text if word not in english_stopwords]
text_without_stopword = filtre_stop(word_tokenize(text_sample, language="english"))
text_without_stopword
```

#### Lemmatisation

La lemmatisation est le processus qui consiste à regrouper les différentes formes infléchies d'un mot afin qu'elles puissent être analysées comme un seul élément. La lemmatisation nltk ne supporte pas la langue française, mais si nous devons l'utiliser pour le français, nous devrions écrire ces deux lignes :

<code>pip install git+https://github.com/ClaudeCoulombe/FrenchLefffLemmatizer.git</code><br>
<code>from french_lefff_lemmatizer.french_lefff_lemmatizer import FrenchLefffLemmatizer</code><br>
<code>lemmatizer = FrenchLefffLemmatizer()</code>

Exemples :
- drones -> drone
- feet -> foot
- oranges -> orange

Au lieu de cela, nous allons utiliser la bibliothèque de lemmatiseurs de nltk.

```python
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("drones")
```

```python
# Lemmatiser une liste de mots et la join
lemmatized_output = [lemmatizer.lemmatize(word) for word in text_without_stopword]
lemmatized_output
```

On a ici utilisé des étapes de nettoyage classique de texte mais en réalité, il est parfois utile de conserver le texte brut quand on est pas dans une recherche du sens d'un document mais de phrases dans leur ensemble puisqu'on cherche le lien entre les différents mots. Par exemple "Ce n'était pas bon", on ne prendrait en compte que le mot "bon" alors qu'il y a le mot "pas" devant.

#### Application sur le jeu de données

Nous allons mettre tous les mots de tous les commentaires dans une liste pour pouvoir les analyser.

```python
stopwords.words('english')
```

```python
english_stopwords = stopwords.words('english')
english_stopwords.remove("not", "against", 'no', 'nor', "don't", "c") # Rajouter toutes les négations
```

```python
cleaned_data = []

for review in tqdm(bad_reviews.text, total=len(bad_reviews)):
    # Suppression de la ponctuation et des espaces
    # '\w+' correspond à tout caractère égal à [a-zA-Z0-9_]
    clean_text = []

    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(review) # Renvoie une liste pour chaque mot
    # On en profite au passage pour filtrer les mots dont la longueur est inférieure à 2
    for token in tokens:
        if len(token)>2:
            clean_text.append(token.lower())

    filtre_stop = lambda text: [word for word in clean_text if word not in english_stopwords]
    text_without_stopword = filtre_stop(word_tokenize(text_sample, language="english"))
            
    # Lemmatiser une liste de mots et la join
    lemmatized_output = [lemmatizer.lemmatize(word) for word in text_without_stopword]
    cleaned_data += lemmatized_output
```

#### Dispersion des mots les plus courants

```python
def plot_dispersion(words: list):
    
    text = [word.split() for word in words]
    # Tri de la liste par le mot le plus courant de la liste
    target_words = [item for items, c in Counter(words).most_common() for item in [items] * c]
    
    # Suppression du doublon et conservation de la première occurrence
    unique = []
    target_words = [unique.append(item) for item in target_words if item not in unique]
    target_words = unique[:15]

    # Plot
    visualizer = DispersionPlot(target_words, color='red')
    fig, ax = plt.subplots(figsize=(15, 10))
    visualizer.fit(text)
    visualizer.show()
```

```python
def sent_to_words(sentences):
    """Returns a list of words. Each word is in a list so it returns a
    list of multiple lists (one list for each word).
    """
    tokenized_sentence = tokenizer.tokenize(sentences)
    for token in tqdm(tokenized_sentence, desc="processing text"):
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(token), deacc=True))
```

```python
def get_city_data(city_name):
    """
    Requête la base de donnée de yelp grâce à son api. Les informations
    récupéreées sont :
        - les noms des restaurants
        - les notes des restaurants
        - les commentaires des restaurants
        - les photos des restaurants
        - les ids des restaurants
        - les ids des commentaires
    
    Voici les différentes étapes de récolte des données :
        1) On récupère dans un premier temps l'id des restaurants situés dans
           la ville donnée en paramètre.
           
        2) Une fois les ids récoltés, on récupère les commentaires ainsi que
           les photos.
           
        3) Enregistrement des données dans trois fichiers csv : un fichier
           pour les commentaires et un fichier pour les photos dans le dossier
           "data/raw". Et un fichier regroupant les deux tables dans "data".
           
           Le nom du fichier prend en compte le nom de la ville. Par exemple
           pour city_name = "Paris", le nom du fichier pour les commentaires
           sera "Paris_yelp_reviews.csv" et celui pour les photos :
           "Paris_yelp_photos.csv".
           
    La fonction retourne une DataFrame mergée de deux DataFrames
    (reviews et photos).
    
    Dans la dataframe on retrouve en colonne :
        - 'review_id'
        - 'business_id',
        - 'rating'
        - 'text'
        - 'photos'
        - 'name'
    """
    
    # Définir le point de terminaison, la clé api et l'en-tête
    url = 'https://api.yelp.com/v3/businesses/search'
    yelp_api_key = os.getenv("API_KEY")
    headers = {'Authorization': f'Bearer {api_key}'}

    # ID des restaurants
    ids = []
    names = []
    photos = []

    for i in tqdm(range(10), desc=f"Récupération des ids des restaurants de {city_name}"):
        # Définir les paramètres de recherche des restaurants
        params = {
            'term': 'restaurant',
            'limit': 50,
            'location': city_name,
            'offset': i * 50
        }

        # Faire la requête
        response = requests.get(url=url, headers=headers, params=params)
        # reponse_txt = response_b.text

        # Enregistrement des IDs
        ids += [business['id'] for business in response.json()['businesses']]

    # Point de terminaison des commentaires
    url_review = "https://api.yelp.com/v3/businesses/"

    # Enregistrement des données dans une dataframe
    reviews = pd.DataFrame(columns=['review_id', 'business_id', 'rating', 'text'])
    pics = pd.DataFrame(columns=['business_id', 'name', 'photos'])

    # Récupération des commentaires
    review_count = 0
    for id in tqdm(ids, desc="Récupération des commentaires"):
        url = url_review + str(id) + "/reviews"
        response = requests.get(url=url,
                                headers=headers,
                                params={'locale': 'fr_FR'})
        for review in response.json()['reviews']:
            reviews.loc[review_count] = [
                review['id'], id, review['rating'], review['text']
            ]
            review_count += 1

    # Récupération des photos
    pic_count = 0
    for id in tqdm(ids, desc="Récupération des photos"):
        url = url_review + str(id)
        response = requests.get(url=url, headers=headers)
        for pic in response.json():
            pics.loc[pic_count] = [
                id, response.json()['name'],
                response.json()['photos']
            ]
            pic_count += 1
    
    # Données sous un seul tableau
    merged_data = reviews.merge(pics, how='inner', on="business_id")
    merged_data = merged_data.drop_duplicates(subset=['review_id'])
    
    # Enregistrement dans un fichier CSV
    reviews.to_csv(f"data/raw/{city_name}_yelp_reviews.csv", index=False)
    pics.to_csv(f"data/raw/{city_name}_yelp_photos.csv", index=False)
    merged_data.to_csv(f"data/{city_name}_yelp_data.csv", index=False)
    
    return merged_data
```
